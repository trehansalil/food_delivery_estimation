```python
import random

class Ship:
    def __init__(self, name, length):
        self.name = name
        self.length = length
        self.positions = []

    def place(self, start_row, start_col, orientation, board):
        # Logic to set ship positions based on orientation and verify it fits
        offsets = {'right': (0, 1), 'down': (1, 0)}
        for i in range(self.length):
            dx, dy = offsets[orientation]
            x, y = start_row + dx * i, start_col + dy * i
            if x >= board.size or y >= board.size or board.grid[x][y] != '.':
                return False  # Invalid placement
            self.positions.append((x, y))
        for x, y in self.positions:
            board.grid[x][y] = 'O'  # O denotes a ship part
        return True

class Board:
    def __init__(self, size=10):
        self.size = size
        self.grid = [['.' for _ in range(size)] for _ in range(size)]
        self.ships = []

    def print_board(self, reveal_ships=False):
        for row in self.grid:
            print(' '.join(row))
        print()

    def place_ships_randomly(self):
        for _ in range(2):  # Simplification for example; adjust for actual game logic
            ship = Ship("Destroyer", 3)
            placed = False
            while not placed:
                row = random.randint(0, self.size - 1)
                col = random.randint(0, self.size - 1)
                orientation = random.choice(['right', 'down'])
                placed = ship.place(row, col, orientation, self)

class Game:
    def __init__(self):
        self.board = Board()
        self.turns = 0

    def play(self):
        # Main game loop; simplified for example
        self.board.place_ships_randomly()
        self.board.print_board()

# Example usage
game = Game()
game.play()
```
